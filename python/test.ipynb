```python
import requests
import json
import pprint
import pandas as pd
import datetime
```

```python
api = 'dcf3ea2273d83ed41036182c6e5a60e9 '
url = 'https://api.chatwork.com/v2'
roomID = '192770919'

headers = {'X-ChatWorkToken': api}
target = '{}/rooms/{}/messages?force={}'.format(url, roomID, 1)
response = requests.get(target, headers = headers)
pprint.pprint(response)

dict_data = json.loads(response.content)
```

```python
f = open("output.json", "w")
json.dump(dict_data, f, indent=4, sort_keys=True, separators=(',', ': '))
```

```python
user_name = dict_data[0]["account"]["name"]
print(user_name)
dt = datetime.datetime.fromtimestamp(dict_data[0]["send_time"])
print(dt)
pprint.pprint(dict_data[3]["body"].split('\n'))

print(int(dt.day))

# ヤス　トモ
# 2020-06-13 16:24:19
# ['1テスト', '2テスト', '3テスト', '4テスト']
# 13

```
